```python
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
from urllib.request import urlopen
```

```python
url ='https://wuzzuf.net/search/jobs/?q=Data+analysis+&a=hpb'
```

```python
client = urlopen(url)
```

```python
html=client.read()
```

```python
client.close()
```

```python
soup=bs(html, 'html.parser')
```

```python
soup
```

```python
cons=soup.find_all('div',{'class':'css-pkv5jc'})
```

```python
file = open('wuzzuf-dataAnalysis.csv','w')
header="Job_title,Company_Name,Job_Type\n"
file.write(header)
```

```python
for con in cons:
    jobtitle=con.findAll('h2',{'class':'css-m604qf'})
    job_title=jobtitle[0].text
    companyName=con.findAll('a',{'class':'css-17s97q8'})
    company_name=companyName[0].text
    jobtype=con.findAll('span',{'class':'css-1ve4b75 eoyjyou0'})
    jop_type=jobtype[0].text
    file.write(job_title+', '+company_name+', '+jop_type+'\n')
file.close()
```
